```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Part 1 - Implement Basic Data Structures using Numpy, Pandas
# 1. Lists
list1 = [2, 4, 7, 6, 8]
l = pd.Series(list1)
# 2. Arrays
array1 = np.array([1, 5, 7, 4, 5])
# 3. Identify their type using type()
print("Type of list1:", l)
print("Type of array1:", array1)
# 4. Perform mathematical operations
print("Multiplication:", array1 * 2)
print("Division:", array1 / 2)
print("Power of:", array1 ** 2)
# 5. Combine text with Numpy function
print("Addition of Two:", np.add(array1, array1))
# 6. Implement np.sin(), log(), log2(), np.exp()
print("Sin:", np.sin(array1))
print("Log:", np.log(array1))
print("Log2:", np.log2(array1))
print("Exp:", np.exp(array1))
```

```python
people = ['sathwik', 'nikhil', 'jupaka', 'sandy']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# Scatter Plot
plt.scatter(age, height)
plt.title('Age vs Height')
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()

# Bar Chart
plt.bar(people, weight)
plt.title('Weight of People')
plt.xlabel('People')
plt.ylabel('Weight')
plt.show()

# Histogram
plt.hist(age, bins=3)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import requests

# URL of the text file
url = "https://example.com/your_text_file.txt"

# Fetch the content from the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract the text content
    text_content = response.text

    # Display the text content
    print("Text from URL:")
    print(text_content)
else:
    print("Failed to fetch data from the URL.")
```
